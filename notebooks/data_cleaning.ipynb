```python
import pandas as pd
import re
import numpy as np
```

```python
PERSON_NAME = ''
INPUT_PATH = ''
OUTPUT_PATH = ''
```

```python
# Get raw data
data = open(INPUT_PATH).read()
raw_data = data.lower().split("\n")
```

```python
# Create a dataframe with data organized by timestamp, name and message
timestamps_array = []
messages_array = []
names_array = []
df = pd.DataFrame(None, columns=['timestamp', 'name', 'message'])
for i, line in enumerate(raw_data[1:]):
    result = re.match(r"(.*) - (.*): (.*)", line)
    if result:
        timestamps_array.append(result.group(1))
        names_array.append(result.group(2))
        messages_array.append(result.group(3))

df = pd.DataFrame({'timestamp': timestamps_array, 'name': names_array, 'message': messages_array })
df
```

```python
df.message.map(lambda x: len(x.split(" "))).max()
mask = df['message'].str.len() <= 100
df_filtered = df.loc[mask]
df_filtered.reset_index()
```

```python
import csv
df_sentences_by_person = df_filtered[df_filtered.name==PERSON_NAME]
df_corpus = df_sentences_by_person['message']
df_corpus.to_csv(OUTPUT_PATH, header=None, index=None,  quoting=csv.QUOTE_NONE, escapechar=" ", sep='\t', mode='a')
```
